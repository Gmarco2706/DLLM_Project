```python
#caricamento del dataset
import pandas as pd


df = pd.read_csv('/Users/marcogreco/DLLM_ProjectBiometria/DLLM_Project/Dataset/heloc_dataset.csv')
```

```python
#standardizzazione con NaN

#Valori mancanti
uncertain_values = [-7, -8, -9]

# Contare i valori prima della sostituzione
print("\nconteggio pre sostituzione:")
total_before = 0
for val in uncertain_values:
    count = (df == val).sum().sum()  
    if count > 0:
        print(f"Valori {val}: {count:,}")
        total_before += count

if total_before == 0:
    print("Nessun valore -7, -8 o -9 trovato")
else:
    print(f"\nTotale valori incerti: {total_before:,}")

# Sostituzione con NaN
df_cleaned = df.replace(uncertain_values, pd.NA)

# Verifica della sostituzione
print("\n conteggio post sostituzione:")
total_after = 0
for val in uncertain_values:
    count = (df_cleaned == val).sum().sum()
    if count > 0:
        total_after += count

if total_after == 0:
    print(f"valori sostituiti con successo")
else:
    print(f"valori non sostituiti: {total_after:,}")

# NaN totali dopo pulizia
total_nan_before = df.isnull().sum().sum()
total_nan_after = df_cleaned.isnull().sum().sum()

print(f"\nValori NaN prima: {total_nan_before:,}")
print(f"Valori NaN dopo: {total_nan_after:,}")
print(f"Incremento NaN: {total_nan_after - total_nan_before:,}")

#salvataggio del dataset pulito
df_cleaned.to_csv('/Users/marcogreco/DLLM_ProjectBiometria/DLLM_Project/Dataset/heloc_cleaned.csv', index=False)
```

```python
#Creazione del dataset pulito senza NaN
df_no_na = df_cleaned.dropna()
df_no_na.to_csv('/Users/marcogreco/DLLM_ProjectBiometria/DLLM_Project/Dataset/heloc_no_na.csv', index=False)
```

```python
#Creazione csv con riepilogo finale


#conteggio colonne numeriche, le colonne numeriche sono state calcolate su object perchè pd.NA ha converitto tutti i formati in object
numeric_cols = (df_no_na.select_dtypes(include=['object']).shape[1]) - 1
#conteggio colonne categoriche, solo una che sarebbe la variabile target RiskPerformance
categorical_cols = 1

original_rows = len(df)
final_rows = len(df_no_na)
n_removed_rows = original_rows - final_rows

#dizionario dei dati
dati = {
    'Dataset': ['heloc_dataset', 'heloc_cleaned', 'heloc_no_na'],
    'N. features numeriche': [numeric_cols, numeric_cols, numeric_cols],
    'N. features categoriche': [categorical_cols, categorical_cols, categorical_cols],
    'N. totale features': [numeric_cols + categorical_cols, numeric_cols + categorical_cols, numeric_cols + categorical_cols],
    'N. campioni originali': [original_rows, original_rows, original_rows],
    'N. campioni post pulizia': [original_rows, original_rows, final_rows],
    'N. righe eliminate': [0, 0, n_removed_rows]
}
#conversione a dataset e salvataggio
riepilogo_dataset = pd.DataFrame(dati)
riepilogo_dataset.to_csv('/Users/marcogreco/DLLM_ProjectBiometria/DLLM_Project/Dataset/riepilogo_dataset.csv', index=False)
```
